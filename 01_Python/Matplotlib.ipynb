# sin, cos 함수

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 2*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# 그래프 그리기
plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label="cos", linestyle="--")
plt.title("sin & cos")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
