# Step Function

```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return np.array(x > 0, dtype=int)

x = np.arange(-10.0, 10.1, 0.1)
y = step_function(x)

plt.plot(x, y)
plt.title("Step Function")
plt.show()
```

# Sigmoid Function

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-10.0, 10.1, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid Function")
plt.show()
```

# ReLU Function

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

x = np.arange(-10.0, 10.1, 0.1)
y = relu(x)

plt.plot(x, y)
plt.title("ReLU Function")
plt.show()
```
